Import all the required packages

```python
#Pandas data is short for panel data that help to manipulate data in python
import pandas as pd
#Numpy is help to generate random numbers and do other many things
import numpy as np
```

```python
#Import the Bread Cancer dataset
from sklearn.datasets import load_breast_cancer
#Load the dataset
data = load_breast_cancer()
#print the data
print(data)

```

```python
#View the data in a DataFrame
df = pd.DataFrame(data.data, columns=data.feature_names)
print(df)
```

```python
#View the traget class data
df['target'] = data.target
```

```python
#Print the target class data
print(df['target'])
```

```python
#Can view the top 5 data in the dataset
df.head()
```

Model Training

```python
#Get values for the X
#Drop out the "target" feature and use remaining features for tarin the model
X = df.drop(['target'], axis =1 )
X.head()
```

```python
#Get values for y
y= df['target']
y.head()
```

Create the traing and testing data

```python
#Split the data into Train and Test by using the train_test_split
from sklearn.model_selection import train_test_split
```

```python
#Split the data, 80% for  training data set and 20% test dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 20)
```

```python
#Check the size of the training set
X_train.shape
```

```python
#Check the Size of the testing data
X_test.shape
```

```python
#Check the size of the training set
y_train.shape
```

```python
#Check the Size of the testing data
y_test.shape
```

```python
#Print the X_train dataset
print(X_train)
```

```python
#Print the X_test data set
print(X_test)
```

```python
#Print the y_tarin data set
print(y_train)
```

```python
 #Print the y_test data set
 print(y_test)
```

Building the Model

```python
#Import the Support Vector Machine (SVM) model
from sklearn.svm import SVC
```

```python
model = SVC()
```

```python
#Fit the Model using the fit() method by pssing the X_train and y_train
model.fit(X_train,y_train)
```

```python
#Make predictions using the test data
pred = model.predict(X_test)
```

```python
#Visualizing the Performance
from sklearn.metrics import classification_report, confusion_matrix
```

```python
#Creat a classification report based on the test data
print(classification_report(y_test, pred))
```

```python
#This model predict that 48 healthy patient have cancer.
```

```python
#Confusion matrix reprsent the performance of the model to classify labels
cm = np.array(confusion_matrix(y_test, pred, labels=[1,0]))
confusion = pd.DataFrame(cm, index=['is_cancer', 'is_healthy'],
                         columns=['predicted_cancer','predicted_healthy'])
confusion
```

Improve the Model Performance by Normalization

```python
#IMPORT PREPROCESSING METHODS
from sklearn import preprocessing
```

```python
min_max_scaler = preprocessing.MinMaxScaler()
```

```python
#Normalization and Fit transform to X_tain data and X_test data
X_train_scaled = min_max_scaler.fit_transform(X_train)
X_test_scaled = min_max_scaler.fit_transform(X_test)
```

```python
#Fit the Model using the fit() method by pssing the X_train_scaled and y_train
model.fit(X_train_scaled,y_train)
```

```python
#Make predictions using the test data
y_predict = model.predict(X_test_scaled)
```

```python
#Creat a classification report based on the test data
print(classification_report(y_test, y_predict))
```

```python
#Confusion matrix reprsent the performance of the model to classify labels
cm = np.array(confusion_matrix(y_test, y_predict, labels=[1,0]))
confusion = pd.DataFrame(cm, index=['is_cancer', 'is_healthy'],
                         columns=['predicted_cancer','predicted_healthy'])
confusion
```
